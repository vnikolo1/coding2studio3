```python
import pandas as pd;
import numpy as np;
```

```python
messy_data = pd.read_csv("C:/Users/valni/AppData/Local/Programs/DTSCProjects/messy_dataset.csv")
messy_data.head()
messy_data.info()
messy_data.describe()
messy_data.isnull().sum()
```

```python
for x in messy_data.index:
    if messy_data.loc[x, "Age"] > 100 :
        messy_data.loc[x, "Age"] = 100
    if messy_data.loc[x, "Age"] < 0 :
        messy_data.loc[x, "Age"] = 0
```

```python
x = messy_data["Age"].mean()
messy_data.fillna({"Age" : x}, inplace=True)
```

```python
print (messy_data)
```

```python
messy_data.dropna(subset=['LastPurchaseDate'], inplace=True)
messy_data.head()
```

```python
messy_data['LastPurchaseDate'] = messy_data['LastPurchaseDate'].astype('datetime64[ns]')
messy_data.dropna(subset=['LastPurchaseDate'], inplace=True)
print (messy_data)
```

```python
messy_data.fillna({"Purchases" : 0}, inplace=True)
for x in messy_data.index:
    if messy_data.loc[x, "Purchases"] == 0 :
        messy_data.loc[x, "Purchases"] = 1

print (messy_data)
```

```python
for x in messy_data.index:
    if messy_data.loc[x, "Income"] > 100000 :
        messy_data.loc[x, "Income"] = 100000

fill = messy_data['Income'].mean()
messy_data.fillna({"Income" : fill}, inplace=True)
print (messy_data)
```

```python
messy_data.to_csv(index=False)
```
